```python
import matplotlib.animation as animation

# Generate some sample data
np.random.seed(0)
dates = pd.date_range('2023-01-01', periods=100)
data = np.random.randn(100).cumsum()

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Value': data})

# Initialize the figure and axis
fig, ax = plt.subplots()
line, = ax.plot(df['Date'], df['Value'], '-')

# Function to update the plot
def update(num, df, line):
    line.set_data(df['Date'][:num], df['Value'][:num])
    ax.set_xlim(df['Date'].min(), df['Date'].max())
    ax.set_ylim(df['Value'].min(), df['Value'].max())
    return line,

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=len(df), fargs=[df, line], interval=1, repeat=False)

# Display the animation
plt.show()
```

```python
from matplotlib.widgets import Slider
amp = 1.0
f = 1.5
fig,ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)

t = np.linspace(0.0,2*np.pi,100)
s = amp * (np.sin(f * t))
l, = ax.plot(t,s,lw=2)
ax_slider = plt.axes([0.25,0.1,0.65,0.03],facecolor='lightgoldenrodyellow')
slider = Slider(ax_slider,label='Frequency',valmin=0.1,valmax=5.0,valinit=f)

def updater(val):
    f = slider.val
    l.set_ydata(amp * np.sin(f * t))
    fig.canvas.draw_idle()
slider.on_changed(updater)
```

# IPYWidgets

```python
import ipywidgets
```

```python
def plotter(w=1):
    x = np.random.uniform(0,5,size=100)
    y = 2 * x + w * np.random.normal(size=100)
    plt.plot(x,y)
ipywidgets.interact(plotter,w=(0,10,0.5))
```

```python
from sklearn.datasets import make_moons
def moon_scatter(n_samples=100,noise=0.1):
    moons = make_moons(n_samples=100,noise=noise,random_state=42)
    X,y = moons[0],moons[1]
    plt.scatter(X[:,0],X[:,1],c=y)
ipywidgets.interact(moon_scatter,n_samples=[200,500,1000],noise=(0,2,0.001))
```

```python
def plot_sin(start=0,end=30,factor=1,grid=False,plot_cos=False):
    X = np.linspace(start,end,(end-start)*10)
    y = np.sin(X) * factor
    plt.grid(grid)
    plt.plot(X,y)
    if plot_cos:
        y = np.cos(X)
        plt.plot(X,y)

ipywidgets.interact(plot_sin,start=(0,10,1),end=(20,50,1),factor=(0,5,0.1),grid=False)
```

```python
import math
def sigmoid(z = 0):
    X = np.linspace(-5,5,100)
    y = 1/(1+np.exp(-X))
    y1 = 1/(1+math.exp(-z))
    plt.plot(X,y,c='b')
    plt.scatter(z,y1,c='r')
    plt.plot([z,z],[0,y1],c='g')
    plt.plot([-5,z],[y1,y1],c='g')
ipywidgets.interact(sigmoid,z=(0,1,0.00001))
```

```python
def hist_plotter(mu=0,sigma=1,n=100,bins=10,color='red'):
    plt.xlim(-20,20)
    X = np.random.normal(mu,sigma,n)
    plt.hist(X,bins=bins,color=color)
#ipywidgets.interact(hist_plotter,mu=(-10,10,0.5),sigma=(0,10,0.1),n=(10,1000,1),bins=1,100,1,color=['red','blue','green'])
ipywidgets.interact(hist_plotter,mu=(-10,10,0.5),sigma=(0,10,0.1),n=(10,1000,1),bins=(1,100,1),color=['red','green','blue'])
```
